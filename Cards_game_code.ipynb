```python
import statistics 


dict={}
def play(games=1):
    j=0
    li=[]
    while j<games: 
        no_of_ties=0
        winning_pile=[]
        cards=[j for i in range(1,5) for j in range(1,14)]         # deck of cards using list comprehension
        np.random.shuffle(cards)                                   # shuffle the deck of cards
        player1=[]
        player2=[]
        k=0
        while k <len(cards):        
            player1.append(cards[k])                              # distribute cards to the players alternatively
            if (k+1<len(cards)):
                 player2.append(cards[k+1])  
            k=k+2
        initial_total1=sum(player1)                           #initial sum of cards of player1
        initial_total2=sum(player2)                           #initial sum of cards of player2
        discardpile1=[]                                       #discard pile of player1
        discardpile2=[]                                       #discard pile of player2
        i=0
        while i<len(player1):                                 # using while loop to iterate through players cards
            sum1=0                                            # initialize sum1 to 0, sum1 is sum of 2 cards played by player1
            sum2=0
            p1=[]                                             # initialize p1 i.e cards played by player1 
            p2=[]                                             # initialize p2 i.e cards played by player2
            p1.append(player1[i])                             #append cards to both lists 
            p1.append(player1[i+1])
            p2.append(player2[i])
            p2.append(player2[i+1])
            sum1=sum(p1)                                        #sum to find which player has greater sum of cards
            sum2=sum(p2)
            if(sum1>sum2):                                     # check if sum1 is greater than sum2                              
                discardpile1.extend(p1)                        # add cards to discardpile of player1
                discardpile1.extend(p2)
                winning_pile.append(sum1)
                p1=[]                                             #empty p1 and p2
                p2=[]
                i=i+2                                         
            elif(sum2>sum1):                                # check if sum2 is greater than sum1                          
                discardpile2.extend(p1)                     # add cards to discardpile of player2
                discardpile2.extend(p2)
                winning_pile.append(sum2)
                p1=[]                                                 #empty p1 and p2
                p2=[]
                i=i+2
            elif(sum1==sum2):
                while(sum1==sum2 and i+2<len(player1)):       # while loop with condition the player have cards and sum is equal for both players 
                    no_of_ties+=1
                    i=i+2
                    p1.append(player1[i])                      # append and store respective players cards in two list until tie breaks or players cards exhaust 
                    p1.append(player1[i+1])
                    p2.append(player2[i])
                    p2.append(player2[i+1])
                    sum1=sum(p1)                                   # add sum of new cards to existing sum1
                    sum2=sum(p2)                                   # add sum of new cards to existing sum2
                    if(i+1==(len(player1)-1) and sum1==sum2):
                        no_of_ties+=1
                i=i+2
                if(sum1>sum2):                                  # check if player1 has greater final score with respect to player2, player1 won the round    
                    discardpile1.extend(p1)                     # empty p1 list to discardpile of player1
                    discardpile1.extend(p2)                     # empty p2 list to discardpile of player1
                    
                    winning_pile.append(sum1)                   #calculating the winning total after tie is broken including the previous tied cards
                    p1=[]                                       # empty p1 and p2
                    p2=[]
                elif(sum2>sum1):                                # check if player1 has greater final score with respect to player2, player1 won the round
                    discardpile2.extend(p1)                     # empty p1 list to discardpile of player2                
                    discardpile2.extend(p2)                     # empty p1 list to discardpile of player2
                    winning_pile.append(sum2)                   #calculating the winning total after tie is broken including the previous tied cards
                    p1=[]
                    p2=[]
                else:
                    discardpile1.extend(p1)
                    discardpile2.extend(p2)
                    p1=[]
                    p2=[]
        score1=sum(discardpile1)                                 # final score of player1 is sum of elements in discardpile1 
        score2=sum(discardpile2)                                 # final score of player2 is sum of elements in discardpile2
        max_round_total=max(winning_pile)                        # maxinmum winning total
        li=[initial_total1,initial_total2,no_of_ties,max_round_total,score1,score2]
        dict[j+1]=li                                             # assign value in dict with key as game number 
        
        j=j+1
    return dict
```

```python
def result(x):                                                    # define function result
    if x['score1']>x['score2']:                                  # if final score of player 1 is greater
        return "Player 1 wins"                                   # then Player 1 wins
    elif x['score1']<x['score2']:                                # if final score of player 2 is greater
        return "Player 2 wins"                                   # then Player 2 wins
    else:                                                        # else there is a tie
        return "Tie game"
    
dfObj['outcome']=dfObj.apply(lambda x : result(x),axis=1)             # adding the column outcome to the dataframe
dfObj.head()
```

### Exercise 3.2

```python
def newcol_win(x):                                   # define function newcol_win
    if x['score1']>x['score2']:                      # if final score of player 1 is greater
        return x['score1']                           # return score of player 1 
    elif x['score1']<x['score2']:                    # if final score of player 2 is greater
        return x['score2']                           # return score of player 2
    else:                                            # in case of tie return np.NaN
        return np.NaN

dfObj['winning_total']=dfObj.apply(lambda x : newcol_win(x),axis=1)  # adding the column winning_total to the dataframe
dfObj.head()
```

```python
def newcol_lose(x):                                        # define function newcol_lose
    if x['score1']>x['score2']:                            # if final score of player 1 is greater
        return x['score2']                                 # return score of player 2
    elif x['score1']<x['score2']:                          # if final score of player 2 is greater
        return x['score1']                                 # return score of player 1
    else:                                                  # in case of tie return np.NaN
        return np.NaN
dfObj['losing_total']=dfObj.apply(lambda x : newcol_lose(x),axis=1) # adding the column losing_total to the dataframe
dfObj.head()
```
